# PCA vino

La reducción de dimensionalidad es una técnica clave en el análisis de datos, que permite simplificar conjuntos complejos y facilitar su interpretación. El Análisis de Componentes Principales (PCA) es uno de los métodos más utilizados para lograr esto, transformando variables correlacionadas en un conjunto de variables no correlacionadas, denominadas componentes principales.
El Wine Dataset, que incluye características químicas de diversos tipos de vino, es ideal para aplicar PCA, ya que permite identificar patrones y relaciones entre las variables. Este reporte presenta la aplicación de PCA en este conjunto de datos, abarcando la normalización de los datos, el cálculo de la matriz de covarianza y la obtención de las componentes principales, junto con visualizaciones que ilustran los resultados.

# PCA Wine Dataset

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Carga de datos del Wine Dataset

```python
from sklearn.datasets import load_wine
data = load_wine()
X = data.data
y = data.target
```

## Normalización de los datos

```python
X_scaled = StandardScaler().fit_transform(X)
```

## Aplicación de PCA (reducimos a 2 componentes para visualizar)

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

##  Visualización de los datos transformados

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='plasma')
plt.xlabel('Primer componente principal')
plt.ylabel('Segundo componente principal')
plt.title('PCA aplicado al Wine Dataset')
plt.show()
```
